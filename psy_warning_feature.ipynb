```python
import pandas as pd
import numpy as np
```

```python
#数据路径

path='F:/weiwei/xue_qian_yu_jing/xqyj_data'

#学生学业预警数据

#2015-2016
psy_warnings_data1=pd.read_csv(path+'/psychology_warning_results(2015-2016).csv',encoding='gbk')
#2016-2017
psy_warnings_data2=pd.read_csv(path+'/psychology_warning_results(2016-2017).csv',encoding='gbk')
```

```python
psy_warnings_data1.head()
```

```python
psy_warnings_data1.columns
```

```python
#删除无用列

X=[x for x in psy_warnings_data1.columns if x not in ['sno','desp']]
psy_warnings_data1=psy_warnings_data1.drop(X,axis=1)
psy_warnings_data2=psy_warnings_data2.drop(X,axis=1)
```

```python
#连接两表

psy_warnings_data=pd.concat([psy_warnings_data1,psy_warnings_data2],ignore_index=True)
```

```python
#desp这一列置0
psy_warnings_data['desp']=1
```

```python
#把desp列名换为is_psy_warning
psy_warnings_data.rename(columns={'desp':'is_psy_warning'},inplace=True)
```

```python
#保存数据
psy_warnings_data.to_csv('psy_data/psy.csv',index=False)
```
